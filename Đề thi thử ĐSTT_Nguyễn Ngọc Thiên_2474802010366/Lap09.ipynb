# ĐỀ THI THỬ ĐSTT

```python
#Câu 1: Phép toán cơ bản trên vector và ma trận
import numpy as np
from numpy.linalg import inv, det, matrix_rank, eig, norm, solve
vec_A = np.array([2, -1, 3, 0])
vec_B = np.array([1, 4, -2, 5])
ma_tran_C = np.array([[1, 2, -1],
                      [0, 3, 4],
                      [2, -1, 0]])

# a) A + B và 3A - 2B
tong_vec = vec_A + vec_B
bieuthuc_vec = 3 * vec_A - 2 * vec_B
print("a) A + B =", tong_vec)
print("   3A - 2B =", bieuthuc_vec)

# b) Tích vô hướng và góc giữa 2 vector
tich_vo_huong = np.dot(vec_A, vec_B)
cos_goc = tich_vo_huong / (np.linalg.norm(vec_A) * np.linalg.norm(vec_B))
goc_do = np.degrees(np.arccos(np.clip(cos_goc, -1.0, 1.0)))
print("b) Tích vô hướng A · B =", tich_vo_huong)
print("   Góc giữa A và B (độ):", round(goc_do, 3))

# c) C.T @ C
ct_nhan_c = np.dot(ma_tran_C.T, ma_tran_C)
print("c) C^T * C =\n", ct_nhan_c)

```

```python
# Câu 2: Nhân ma trận, định thức và hạng ma trận
import numpy as np
from numpy.linalg import inv, det, matrix_rank, eig, norm, solve
ma_tran_M = np.array([[1, 2, 3, 0],
                      [4, 5, 6, 1],
                      [7, 8, 9, 2]])
ma_tran_N = np.array([[0, 1, 2],
                      [4, 5, 6],
                      [8, 9, 10],
                      [12, 13, 15]])

# a) Tích M × N
tich_MN = np.dot(ma_tran_M, ma_tran_N)
print("a) M x N =\n", tich_MN)

# b) Định thức M
dinh_thuc_M = det(ma_tran_M) if ma_tran_M.shape[0] == ma_tran_M.shape[1] else "Không khả thi (không vuông)"
print("b) Định thức của M:", dinh_thuc_M)

# c) Hạng của N
hang_N = matrix_rank(ma_tran_N)
print("c) Hạng của N =", hang_N)
```

```python
#Câu 3: Nghịch đảo ma trận và giải hệ phương trình
import numpy as np
from numpy.linalg import inv, det, matrix_rank, eig, norm, solve
ma_tran_P = np.array([[2, 1, 0, 1],
                      [0, 3, -1, 2],
                      [1, 0, 1, 0],
                      [1, 0, 0, 3]])
vec_b3 = np.array([4, 7, 9, 5])

if det(ma_tran_P) != 0:
    nghich_dao_P = inv(ma_tran_P)
    nghiem_he = solve(ma_tran_P, vec_b3)
    kiemtra_b3 = np.dot(ma_tran_P, nghiem_he)
else:
    nghich_dao_P = "Ma trận không khả nghịch"
    nghiem_he = "Không giải được"
    kiemtra_b3 = "Không kiểm tra được"

print("a) Nghịch đảo của P =\n", nghich_dao_P)
print("b) Nghiệm hệ phương trình =", nghiem_he)
print("c) Kiểm tra P·x = b:", kiemtra_b3)
```

```python
#Câu 4: Giá trị riêng, vector riêng và chuẩn hóa
import numpy as np
from numpy.linalg import inv, det, matrix_rank, eig, norm, solve
ma_tran_Q = np.array([[4, -1, 0],
                      [1, 3, -2],
                      [0, 1, 2]])

# a) Giá trị riêng
gia_tri_rieng, vector_rieng = eig(ma_tran_Q)
print("a) Giá trị riêng của Q:", gia_tri_rieng)

# b) Vector riêng
print("b) Vector riêng:\n", vector_rieng)

# c) Chuẩn hóa vector riêng
vector_rieng_chuan = vector_rieng / np.linalg.norm(vector_rieng, axis=0)
print("c) Vector riêng đã chuẩn hóa:\n", vector_rieng_chuan)
```

```python
#Câu 5: Ứng dụng thực tế - Hệ phương trình và phân tích
import numpy as np
from numpy.linalg import inv, det, matrix_rank, eig, norm, solve
ma_tran_R = np.array([[1, -2, 3],
                      [2, 0, 1],
                      [-1, 1, 4]])
vec_b5 = np.array([10, 8, 7])

nghiem_Rx = solve(ma_tran_R, vec_b5)
if det(ma_tran_R) != 0:
    nghich_dao_R = inv(ma_tran_R)
    kiemtra_R = np.dot(nghich_dao_R, vec_b5)
else:
    nghich_dao_R = "Không khả nghịch"
    kiemtra_R = "Không kiểm tra được"

chuan_Frobenius = norm(ma_tran_R, ord='fro')

print("a) Nghiệm hệ R·x = b:", nghiem_Rx)
print("b) Nghịch đảo R =\n", nghich_dao_R)
print("   Kiểm tra lại x:", kiemtra_R)
print("c) Chuẩn Frobenius của R =", chuan_Frobenius)
```
